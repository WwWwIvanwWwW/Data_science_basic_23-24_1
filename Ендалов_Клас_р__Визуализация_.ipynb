```python
fig, ax=plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.gcf().set_size_inches(10, 10)

#1

ax[0,0].plot(x, y)
plt.subplot(2, 2, 1)
plt.xlabel("x")
plt.ylabel("y")
plt.title(label="№1", loc="center")
#2
ax[0,1].scatter(x, y)
plt.subplot(2, 2, 2)
plt.xlabel("x")
plt.ylabel("y")
plt.title(label="№2", loc="center")
#3
ax[1, 0].hist(x)
plt.subplot(2, 2, 3)
plt.xlabel("x")
plt.ylabel("y")
plt.title(label="№3", loc="center")
#4
ax[1, 1].hist(y)
plt.subplot(2, 2, 4)
plt.xlabel("y")
plt.ylabel("x")
plt.title(label="№4", loc="center")
plt.show()
```
